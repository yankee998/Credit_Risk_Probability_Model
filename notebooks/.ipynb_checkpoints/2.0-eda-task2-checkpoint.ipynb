```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
from sklearn.preprocessing import LabelEncoder

# Set style and palette for better visualizations
plt.style.use('seaborn-v0_8-whitegrid')  # Valid style from available list
sns.set_palette("husl")  # Seaborn color palette

# Load the data with absolute paths
data = pd.read_csv('C:\\\\Users\\\\Skyline\\\\Credit Risk Probability Model\\\\data\\\\raw\\\\data.csv')
variable_defs = pd.read_csv('C:\\\\Users\\\\Skyline\\\\Credit Risk Probability Model\\\\data\\\\raw\\\\Xente_Variable_Definitions.csv')

# Display basic info
print("Dataset Shape:", data.shape)
print("\nDataset Info:")
print(data.info())
print("\nVariable Definitions:")
print(variable_defs.head())

# Data preprocessing
data['TransactionStartTime'] = pd.to_datetime(data['TransactionStartTime'])
data['TransactionHour'] = data['TransactionStartTime'].dt.hour
data['TransactionDay'] = data['TransactionStartTime'].dt.day
data['TransactionMonth'] = data['TransactionStartTime'].dt.month
data['TransactionYear'] = data['TransactionStartTime'].dt.year

# Handle categorical variables
le = LabelEncoder()
data['ChannelId'] = le.fit_transform(data['ChannelId'])
data['PricingStrategy'] = le.fit_transform(data['PricingStrategy'])

# Check for missing values
print("\nMissing Values:")
print(data.isnull().sum())

# Visualize target variable distribution
plt.figure(figsize=(8, 6))
sns.countplot(x='FraudResult', data=data)
plt.title('Distribution of Fraud Results (Target Variable)')
plt.xlabel('Fraud Result (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()

# Check class imbalance
fraud_rate = data['FraudResult'].mean() * 100
print(f"\nPercentage of Fraudulent Transactions: {fraud_rate:.2f}%")

# Explore feature relationships
plt.figure(figsize=(10, 8))
numeric_data = data.select_dtypes(include=[np.number])
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap of Numerical Features')
plt.show()

# Fraud by TransactionHour
plt.figure(figsize=(12, 6))
sns.boxplot(x='TransactionHour', y='FraudResult', data=data)
plt.title('Fraud Rate by Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Fraud Result')
plt.show()

# Feature importance analysis (basic)
features = ['TransactionHour', 'PricingStrategy', 'ChannelId', 'Amount']
for feature in features:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x='FraudResult', y=feature, data=data)
    plt.title(f'Distribution of {feature} by Fraud Result')
    plt.xlabel('Fraud Result (0 = No, 1 = Yes)')
    plt.ylabel(feature)
    plt.show()

# Save processed data for modeling
data.to_csv('C:\\\\Users\\\\Skyline\\\\Credit Risk Probability Model\\\\data\\\\processed\\\\task2_eda_data.csv', index=False)
print("\nProcessed data saved to 'C:\\\\Users\\\\Skyline\\\\Credit Risk Probability Model\\\\data\\\\processed\\\\task2_eda_data.csv'")
```
